```python
# the dataset

dogs_df = pd.read_csv('/data/notebook_files/dog_breeds.csv')
```

```python
# Creating function to extract string, split it and convert into integer

def extract_numbers(value, array_index):
    return int(value.split('-')[array_index])

def extract_int(dataframe, column_name, array_index):
    list_of_values = []
    for index,row in dataframe.iterrows():
        value = (row[column_name])
        list_of_values.append(extract_numbers(value,array_index))
        
    return list_of_values
        
 
    
Height_min = extract_int(dogs_df,'Height (in)', 0)
Height_max = extract_int(dogs_df,'Height (in)', 1)
Longevity_min = extract_int(dogs_df,'Longevity (yrs)', 0)
Longevity_max = extract_int(dogs_df,'Longevity (yrs)', 1)
                           
    
dogs_df.insert(4,'Height_min',Height_min, True)
dogs_df.insert(5,'Height_max',Height_max, True)
dogs_df.insert(8,'Longevity_min',Longevity_min, True)
dogs_df.insert(9,'Longevity_max',Longevity_max, True)
```

```python
from collections import Counter

list_of_problems = dogs_df['Common Health Problems'].str.lower().str.split(', ').tolist()

# Flatten the list of health problems
flat_list = [problem for problems in list_of_problems for problem in problems]

# Count the occurrence of each health problem
problem_counts = Counter(flat_list)

# Convert dictionary into dataframe
health_df = pd.DataFrame(list(problem_counts.items()), columns=['Issues','Count'])



# Plot a pie plot with data from new dataframe
bg_color = 'silver'
labels = health_df['Issues']
data = health_df['Count']


font_size= 25



figp1, axp1 = plt.subplots(figsize=(15,15), facecolor=bg_color, subplot_kw=dict(aspect="equal"))

perc = [str(round(e / s * 100., 1)) + '%' for s in (sum(data),) for e in data]
wedges, texts = axp1.pie(data,wedgeprops={"edgecolor" : "black",
                      'linewidth': 2,
                      'antialiased': True}, 
                      startangle=160)

kw = dict(arrowprops=dict(arrowstyle="-"),zorder=0, va="center", fontsize=font_size)




for i, p in enumerate(wedges):
    if float(perc[i][:-1]) >= 0.9:
        ang = (p.theta2 - p.theta1)/2. + p.theta1
        y = np.sin(np.deg2rad(ang))
        x = np.cos(np.deg2rad(ang))
        horizontalalignment = {-1: "right", 1: "left"}[int(np.sign(x))]
        connectionstyle = "angle,angleA=0,angleB={}".format(ang)
        kw["arrowprops"].update({"connectionstyle": connectionstyle})
        axp1.annotate(labels[i] + ' ' + perc[i], xy=(x, y), xytext=(1.1*np.sign(x), 1.1*y),
                      horizontalalignment=horizontalalignment, **kw)




class Circle:
    def __init__(self,c1,c2,color):
        self.c1 = c1
        self.c2 = c2
        self.color = color
    
    def draw(self):
        my_circle=plt.Circle( (0,0), self.c1, color='black')
        my_circle2=plt.Circle( (0,0), self.c2, color=self.color)
        p=plt.gcf()
        p.gca().add_artist(my_circle)
        p.gca().add_artist(my_circle2)

circle = Circle(0.6,0.55,bg_color)
circle.draw()


axp1.set_title('Common Health Problems in Dogs (117 breeds) ', fontdict={'fontsize': 30, 'fontweight': 'bold'}, pad=60)
plt.show()
```

```python
from collections import Counter

list_of_traits = dogs_df['Character Traits'].str.lower().str.split(', ').tolist()

# Flatten the list of health problems
flat_list = [trait for traits in list_of_traits for trait in traits]

# Count the occurrence of each health problem
traits_counts = Counter(flat_list)

# Convert dictionary into dataframe
traits_df = pd.DataFrame(list(traits_counts.items()), columns=['Traits','Count'])


traits_df = traits_df.sort_values(by='Count', ascending=False)

# Plot a pie plot with data from new dataframe

pctdistance=0.90

def my_autopct(pct):
    return f"{pct:.1f}%" if pct > 1.5 else ""


def my_labels(label):
    if label == "":
        return label
    elif traits_df.loc[label, 'Count'] > 8:
        return f"{traits_df.loc[label, 'Traits']} ({traits_df.loc[label,'Count']})"
    else:
        return ""



figp2, axp2 = plt.subplots(figsize=(25,25), facecolor=bg_color, subplot_kw=dict(aspect="equal"))
axp2.pie(traits_df['Count'], autopct=my_autopct,pctdistance=pctdistance,startangle=55,
         wedgeprops={"edgecolor" : "black",
                      'linewidth': 4,
                      'antialiased': True},
         textprops={'color': 'black', 'size': 25,'fontweight': 'bold'},
         labels=traits_df.index.map(my_labels)
         )



axp2.set_title('Character Traits in Dogs (117 breeds) ',fontdict={'fontsize': 30, 'fontweight': 'bold'}, pad=5)
plt.legend(traits_df['Traits'],loc='lower center', ncol=4, bbox_to_anchor=(0.5, -0.14), fontsize=25, edgecolor='black')


circle3 = Circle(0.8,0.79,bg_color)
circle3.draw()


plt.show()
```

```python
# Creating a class for exercise purpose 

class Dogs_details:
    def __init__(self, breed, country_of_origin, fur_color, height_min, height_max, color_of_eye, longevity_min, longevity_max, character_traits, common_health_problems): 
        self.breed = breed
        self.country_of_origin = country_of_origin
        self.fur_color = fur_color
        self.height_min = height_min
        self.height_max = height_max
        self.color_of_eye = color_of_eye
        self.longevity_min = longevity_min
        self.longevity_max = longevity_max
        self.character_traits = character_traits
        self.common_health_problems = common_health_problems

list_of_dogs_details = [
    Dogs_details ("Example_dog", "Poland", "Red", 10, 20, "White", 20, 30, "Intelligent", "Eye problems"),


]
```

```python
def add_row(list_of_dogs):
    t = pd.DataFrame()
    for dog_details in list_of_dogs:
        dogs2 = pd.DataFrame({"Breed":[dog_details.breed],
                            "Country of Origin":[dog_details.country_of_origin], 
                            "Fur Color":[dog_details.fur_color], 
                            "Height_min":[dog_details.height_min],
                            "Height_max":[dog_details.height_max],
                            "Color of Eyes":[dog_details.color_of_eye],
                            "Longevity_min":[dog_details.longevity_min],
                            "Longevity_max":[dog_details.longevity_max],
                            "Character Traits":[dog_details.character_traits],
                            "Common Health Problems":[dog_details.common_health_problems]
                            })
        
        t = t.append(dogs2)
        
    return t
    
```

```python
# Fuction returns the number of dog breeds per country of origin

def extract_breed (dataframe : pd.DataFrame, country, name):
    new_data_frame : pd.DataFrame = dataframe.loc[dataframe[country] == name]
    
    return new_data_frame.shape[0]

list_of_countries = dogs_df['Country of Origin'].drop_duplicates(keep='first').sort_values(ascending=True)

dogs = []


for country in list_of_countries:
    function = extract_breed(dogs_df,'Country of Origin',country)
    dog_di={}
    dog_di['Country of Origin']=country
    dog_di['Result']=function
    dogs.append(dog_di)

breed_per_c_df = pd.DataFrame(dogs, columns=('Country of Origin', 'Result'))

breed_per_c_df
```
