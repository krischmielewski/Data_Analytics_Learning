```python
# I'm adding the dataset

dogs_df = pd.read_csv('/data/notebook_files/dog_breeds.csv')
```

```python
# I'm creating function to extract string, split it and convert into integer

def extract_numbers(value, array_index):
    return int(value.split('-')[array_index])

def extract_int(dataframe, column_name, array_index):
    list_of_values = []
    for index,row in dataframe.iterrows():
        value = (row[column_name])
        list_of_values.append(extract_numbers(value,array_index))
        
    return list_of_values
        
 
    
Height_min = extract_int(dogs_df,'Height (in)', 0)
Height_max = extract_int(dogs_df,'Height (in)', 1)
Longevity_min = extract_int(dogs_df,'Longevity (yrs)', 0)
Longevity_max = extract_int(dogs_df,'Longevity (yrs)', 1)
                           
    
dogs_df.insert(4,'Height_min',Height_min, True)
dogs_df.insert(5,'Height_max',Height_max, True)
dogs_df.insert(8,'Longevity_min',Longevity_min, True)
dogs_df.insert(9,'Longevity_max',Longevity_max, True)
```
