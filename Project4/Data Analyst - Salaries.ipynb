```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import numpy as np
import seaborn as sns
```

```python
salaries_df = pd.read_csv(r'/data/notebook_files/ds.salaries.csv')
```

```python
salaries_df.drop(columns='Unnamed: 0',inplace=True)
```

```python
salaries_df
```

```python
columns = list(salaries_df.columns)
columns
```

```python
salaries_2020_df = salaries_df[salaries_df['work_year']== 2020].drop(columns='work_year')
salaries_2021_df = salaries_df[salaries_df['work_year']== 2021].drop(columns='work_year')
salaries_2022_df = salaries_df[salaries_df['work_year']== 2022].drop(columns='work_year')
```

```python
salaries_2020_df = salaries_2020_df.loc[(salaries_2020_df['employment_type'] == 'FT')]
salaries_2021_df = salaries_2021_df.loc[(salaries_2021_df['employment_type'] == 'FT')]
salaries_2022_df = salaries_2022_df.loc[(salaries_2022_df['employment_type'] == 'FT')]
```

```python
salaries_2020_pivot_df = salaries_2020_df.pivot_table(values='salary_in_usd', index=['job_title'], columns='company_location')

columns = salaries_2020_pivot_df.columns

for column in columns:
    salaries_2020_pivot_df[column] = salaries_2020_pivot_df[column].fillna("0")

salaries_2020_pivot_df   
```

```python
salaries_2020_df['job_title_plus'] = salaries_2020_df['job_title'].astype(str) +" - " + salaries_2020_df['experience_level'] + " - " + salaries_2020_df['company_location'] + " - R:" + salaries_2020_df['remote_ratio'].astype(str)
salaries_2021_df['job_title_plus'] = salaries_2021_df['job_title'].astype(str) +" - " + salaries_2021_df['experience_level'] + " - " + salaries_2021_df['company_location'] + " - R:" + salaries_2021_df['remote_ratio'].astype(str)
salaries_2022_df['job_title_plus'] = salaries_2022_df['job_title'].astype(str) +" - " + salaries_2022_df['experience_level'] + " - " + salaries_2022_df['company_location'] + " - R:" + salaries_2022_df['remote_ratio'].astype(str)
```

```python
fig1, ax1 = plt.subplots(figsize=(15,30),facecolor='black')

salaries_2020_df.sort_values(by="salary_in_usd",ascending=False, inplace=True)

job_title = salaries_2020_df['job_title_plus']
salary = salaries_2020_df['salary_in_usd']


ax1.barh(range(len(job_title)),salary, color="green",edgecolor='black')
ax1.set_title('Full time|International|All EXP levels',fontdict={'fontsize': 30, 'fontweight': 'bold'}, color='white',pad=80)

ax1.set_facecolor('black')
ax1.set_xticks(np.arange(0, max(salary)+50000, 50000))
ax1.set_xticklabels(np.arange(0, max(salary)+50000, 50000), fontsize=20, color='white')
ax1.set_yticks(range(len(job_title)))
ax1.set_yticklabels(job_title,size=20,color='white')
plt.tick_params(axis='x', which='both', top=True, bottom=True, labeltop=True, labelbottom=True)
plt.tick_params(axis='y', which='both',pad=25)
plt.suptitle('"Data" related jobs - Salary in 2020 (USD)',fontsize=40, color='white')
ax1.yaxis.set_label_position("right")
ax1.yaxis.tick_right()

plt.show()
```

```python
salaries_2021_df.sort_values(by='salary_in_usd',inplace=True)

job_title = salaries_2021_df['job_title_plus']
salary = salaries_2021_df['salary_in_usd']
figure(num=None, figsize=(12,54), dpi=80, facecolor='black')

ax2 = sns.barplot(y=job_title, x = salary)
ax2.set_title('"Data" related jobs - Salary in 2021 (USD)\nFull time|International|All EXP levels',
              fontdict={'fontsize': 30, 'fontweight': 'bold'}, color='white',pad=20)
sns.set_style("darkgrid", {"axes.facecolor": ".9"})
sns.set_palette('pastel')
ax2.tick_params(axis='x', colors='white',which='both', top=True, bottom=True, labeltop=True, labelbottom=True)
ax2.tick_params(axis='y', colors='white')
ax2.set_xlabel("")
ax2.set_ylabel("")

plt.show()
```

```python
salaries_2022_df.sort_values(by='salary_in_usd',inplace=True)

job_title = salaries_2022_df['job_title_plus']
salary = salaries_2022_df['salary_in_usd']
figure(num=None, figsize=(12,34), dpi=80, facecolor='black')

ax3 = sns.barplot(y=job_title, x = salary)
ax3.set_title('"Data" related jobs - Salary in 2022 (USD)\nFull time|International|All EXP levels',
              fontdict={'fontsize': 30, 'fontweight': 'bold'}, color='white',pad=20)
sns.set_style("darkgrid", {"axes.facecolor": ".9"})
sns.set_palette('pastel')
ax3.tick_params(axis='x', colors='white', which='both', top=True, bottom=True, labeltop=True, labelbottom=True)
ax3.tick_params(axis='y', colors='white')
ax3.set_xlabel("")
ax3.set_ylabel("")

plt.show()
```

```python
count_emp_type = salaries_df['employment_type'].value_counts()
emp_type = dict(zip(count_emp_type.index, count_emp_type.values))

colors_emp = ["#FFD700", "#d62728", "#9467bd","#E6E6FA"]
explode_emp = (0.3,0.6,0.6,0.6)

plt.subplots(figsize=(10,10))
plt.pie(x=emp_type.values(), labels=emp_type.keys(),autopct='%.0f%%',pctdistance=1.15,labeldistance=1.4,
        startangle=1,wedgeprops={"edgecolor":"k",'linewidth': 1, 'linestyle': 'solid', 'antialiased': True},
        textprops={'color': 'black', 'size': 20,'fontweight': 'bold'},
                  colors=colors_emp,explode=explode_emp)


plt.title('Employment type', fontdict={'fontsize': 30, 'fontweight':'bold',},bbox={'facecolor':'0.8','pad':5})
plt.show()
```

```python
count_rem_rat = salaries_df['remote_ratio'].value_counts()
rem_rat = dict(zip(count_rem_rat.index, count_rem_rat.values))

colors_rem = ["#1f77b4", "#ff7f0e", "#2ca02c"]
plt.subplots(figsize=(10,10))
pie_remote = plt.pie(x=rem_rat.values(), labels=rem_rat.keys(),autopct='%.0f%%', 
        startangle=90,wedgeprops={"edgecolor":"k",'linewidth': 1, 'linestyle': 'solid', 'antialiased': True},
                     textprops={'color': 'black', 'size': 25,'fontweight': 'bold'},
                     colors=colors_rem)
plt.title('Remote working', fontdict={'fontsize': 30, 'fontweight': 'bold'},bbox={'facecolor':'0.8', 'pad':5})

plt.show()
```

```python
count_comp_size = salaries_df['company_size'].value_counts()
comp_size = dict(zip(count_comp_size.index, count_comp_size.values))

colors_comp = ["#FA8072", "#FF0000", "#0000FF"]
plt.subplots(figsize=(10,10))
pie_company = plt.pie(x=comp_size.values(), labels=comp_size.keys(),autopct='%.0f%%', 
        startangle=90,wedgeprops={"edgecolor":"k",'linewidth': 1, 'linestyle': 'solid', 'antialiased': True},
                     textprops={'color': 'black', 'size': 25,'fontweight': 'bold'},
                     colors=colors_comp)
plt.title('Company size', fontdict={'fontsize': 30, 'fontweight': 'bold'},bbox={'facecolor':'0.8', 'pad':5})

plt.show()
```
